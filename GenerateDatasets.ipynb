```python
from sklearn.datasets import make_blobs, make_circles, make_classification, make_moons,make_regression, make_multilabel_classification
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y = make_circles(n_samples=100, shuffle=True, noise=0.1,random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot()
```

```python
X, y = make_moons(n_samples=100, noise=0.1, shuffle=True, random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=3, shuffle=True, random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X, y = make_classification(n_samples=500, n_features=3, n_classes=3, n_informative=3, n_redundant=0, n_repeated=0, random_state=23)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X, y = make_multilabel_classification(n_samples=100, n_features=3, n_classes=2, n_labels=2, random_state=23)
df = pd.concat([pd.DataFrame(X, columns=['x1', 'x2', 'x3']),
                pd.DataFrame(y, columns=['label1', 'label2'])], axis=1)
display(df.head())
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=24,random_state=23)
plt.scatter(X, y)
plt.show()
```
